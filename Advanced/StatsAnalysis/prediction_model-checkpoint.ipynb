```python
import matplotlib.pyplot as plt
def create_corr(df,y,lag_c=30,lag_d=30):
    
    fig,ax=plt.subplots(nrows=2,ncols=1)
    ax[1].xcorr(df[y],df["daily_deaths"],"full",maxlags=lag_d)#[-1]
    #print(g2)
    ax[1].set_title("cross-correlation between daily_deaths and {}".format(y))
    #ax[1].show(False)
    ax[0].xcorr(df[y],df["daily_cases"],"full",maxlags=lag_c)#[-1]  
    ax[0].set_title("cross-correlation between daily_cases and {}".format(y))
    #plt.xcorr
    fig.subplots_adjust(hspace=1)
    fig.show(False)
    return fig
```

ARIMA with lagged exogenous variables 

```python
freq_lag=pd.DataFrame()
l20=["sore throat","symptomatic","coronavirus case","flu","cough","coronavirus outbreak","china","fever","cough","flu"]
l10=["headache","respiratory illness","self isolate","chinavirus"]#,"fever","cough"
l0=["mask","infection","coronavirus case","social distance","coronavirus spread","pandemic","wuhan","corona virus","covid"]#"flu"
for i in df_freq.columns[1:-2]:
    if i in l10:
        freq_lag[i]=df_freq[i].shift(10)
    if i in l20:
        freq_lag[i]=df_freq[i].shift(20)
    if i in l0:
        freq_lag[i]=df_freq[i]
freq_lag.fillna(0,inplace=True)
freq_lag
```

```python
import seaborn as sns
from statsmodels.tsa.arima_model import ARIMA
model=ARIMA(df_freq["daily_cases"][:100],order=(4,1,0),exog=freq_lag[:100][freq_lag.columns]) #df_freq[df_freq.columns[1:-1]][:100])
model=model.fit(trend='nc')#,method='mle')
model.summary()
y_fit=model.predict(start=1,end=100,typ="levels",exog=freq_lag[:100][freq_lag.columns])
y_fit[y_fit<0]=0
y_fit
fit_cor=y_fit.corr(df_freq["daily_cases"][:100])
print("correlation btween training  and fitted response is {}".format(fit_cor))
y_pred=model.predict(100,106,exog=freq_lag[100:][freq_lag.columns],typ='levels')
pred_cor=y_pred.corr(df_freq["daily_cases"][100:])
print("correlation btween test  and predicted response is {}".format(pred_cor))

import matplotlib.pyplot as plt
fig,ax=plt.subplots()
sns.lineplot(x=df["created_at"][:100],y=y_fit,ax=ax)
sns.lineplot(x=df["created_at"][:100],y=df["daily_cases"][:100],ax=ax)
ax.legend(labels=["observed","fitted"])
ax.set_title("Observed Covid cases and Fitted Covid cases")

import matplotlib.pyplot as plt
fig,ax1=plt.subplots()
sns.lineplot(x=df["created_at"][100:],y=y_pred,ax=ax1)
sns.lineplot(x=df["created_at"][100:],y=df["daily_cases"][100:],ax=ax1)
ax1.legend(labels=["test","predicted"])
ax1.set_title("Observed Covid cases and predicted Covid cases")
```
